```python
from biokb_taxtree.db.manager import DbManager
from sqlalchemy import create_engine
from biokb_taxtree.db import models
from pprint import pprint
```

# Query data

Data can be queried by using the session from DbManager. Make sure you have created first a database like described in [import_data.ipynb](import_data.ipynb)

```python
engine = create_engine("sqlite:///test.db")
Session = DbManager(engine).Session
```

## Names

```python
with Session() as session:
    query = session.query(models.Name).filter(
        models.Name.name_txt.like("Achillea millefolium%"),
        models.Name.name_class == "scientific name",
    )
    results: list[models.Name] = query.all()
    # pprint(results)
    print("Number of entries:", len(results))
    first_entry = results[0]
    print(first_entry)
```

## Ranked Lineage

```python
with Session() as session:
    query = session.query(models.RankedLineage).filter(
        models.RankedLineage.tax_id == first_entry.tax_id
    )
    ranked_lineage: models.RankedLineage = query.first()
    print(ranked_lineage)
    print(ranked_lineage.superkingdom)
```

## Node

```python
with Session() as session:
    query = session.query(models.Node).filter(models.Node.tax_id == first_entry.tax_id)
    node: models.Node = query.first()
    print(node.ranked_lineage)
    print(node)
```
